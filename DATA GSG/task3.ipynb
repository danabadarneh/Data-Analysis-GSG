Q1: Calculating the number of deaths in a period


```python

df = pd.DataFrame(data1)
df['death_count'] = pd.to_numeric(df['death_count'].replace('.', 0))

# Merge three columns into one as 'Merged_Column' and convert it to datetime
df['Merged_Column'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str))

# Define the specific time period
start_date = pd.Timestamp('2020-01-01')
end_date = pd.Timestamp('2020-4-01')

# Filter the DataFrame to include only the rows within the desired time period
filtered_df = df[(df['Merged_Column'] >= start_date) & (df['Merged_Column'] < end_date)]

# Calculate the total number of deaths in the time period
total_deaths = filtered_df['death_count'].sum()

print("Total deaths in the time period:", total_deaths)
```

Q2: How many new vaccines are there on this date?

```python

df = pd.DataFrame(data1)
# Merge three columns into one as 'Merged_Column' and convert it to datetime
df['Merged_Column'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str))
# Prompt the user to enter the date
date_input = input("Enter the date (YYYY-MM-DD): ")

# Convert user input to Timestamp object
date = pd.Timestamp(date_input)

# Filter the DataFrame to include only the rows for the specified date
filtered_df = df[df['Merged_Column'].dt.date == date.date()]

# Replace '.' with 0 in the 'new_vaccine_count' column and convert it to numeric type
filtered_df['new_vaccine_count'] = pd.to_numeric(filtered_df['new_vaccine_count'].replace('.', 0))

# Calculate the total new vaccine count on the specified date
total_new_vaccine_count = filtered_df['new_vaccine_count'].sum()

print("Total new vaccine count on", date.date(), ":", total_new_vaccine_count)
```

Q3:How many cases appeared in the city according to the ID number?

```python

df = pd.DataFrame(data1)

# Prompt the user to enter the city ID
city_id = int(input("Enter the city ID: "))

# Filter the DataFrame to include only the rows for the specified city ID
filtered_df = df[df['cityid'] == city_id]

# Replace any non-numeric values in the 'case_count' column with zeros and convert it to numeric type
filtered_df['case_count'] = pd.to_numeric(filtered_df['case_count'], errors='coerce').fillna(0)

# Calculate the total number of cases for the specified city
total_cases_in_city = filtered_df['case_count'].sum()

print("Total cases in city with ID", city_id, ":", total_cases_in_city)


```

```python

df = pd.DataFrame(data1)

# Merge three columns into one as 'Date' and convert it to datetime
df['Date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['death_count'] = pd.to_numeric(df['death_count'], errors='coerce').fillna(0)

# Group by date and calculate total death count for each date
df_grouped = df.groupby('Date')['death_count'].sum().reset_index()

# Create the plot using Plotly Express
fig = px.line(df_grouped, x='Date', y='death_count', title='Number of Deaths Over Time')
fig.show()
```

As shown in the chart, the highest rate of infection with the Corona virus was in the third month of the year 2023, and after that the numbers began to decrease dramatically.
